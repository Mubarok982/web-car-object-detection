```python
from ultralytics import YOLO

# Load YOLOv8 nano model
model = YOLO('yolov8n.pt')
# Training
model.train(
    data='/content/drive/MyDrive/yolo-datasets/vehicle-license/data.yaml',
    epochs=50,
    imgsz=640,
    batch=16,
    name='car-damage-detector'
)
```

```python
from google.colab import files
import pickle
import torch

# Path ke model terbaik Anda
model_path = 'runs/detect/car-damage-detector3/weights/best.pt'
pickle_filename = 'car_damage_model.pkl'

try:
    # Muat model dari file .pt dengan menambahkan weights_only=False
    print("Mencoba memuat model...")
    model = torch.load(model_path, weights_only=False)['model']
    print("Model berhasil dimuat.")

    # Simpan model menggunakan pickle
    with open(pickle_filename, 'wb') as file:
        pickle.dump(model, file)
    print(f"✅ File '{pickle_filename}' berhasil dibuat.")

    # Langsung mulai proses download setelah file berhasil dibuat
    print(f"Memulai proses download untuk '{pickle_filename}'...")
    files.download(pickle_filename)

except Exception as e:
    print(f"Terjadi error: {e}")
```
